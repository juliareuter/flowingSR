```julia
Re = 0
n_neighbors = 60
dataset_prefix = "./data/varying_Re"
arbitrary_name = "Re$(Re)_n$(n_neighbors)_symmetricRotation"

data_matrix = Matrix(CSV.read("$(dataset_prefix)/$(arbitrary_name).csv", DataFrame))

#define an equation with variables
eq = sin.((-2.94.*data_matrix[:,2]))./((1.71.*data_matrix[:,1]).^2.46)
#compute mse
eq = eq .- data_matrix[:,11]
mse = sum(eq.^2)/length(eq)
println("MSE: ", mse)
#compute R^2
R2 = 1 - sum(eq.^2)/sum((data_matrix[:,11].-mean(data_matrix[:,11])).^2)
println("R^2: ", R2)
```

```julia
Re = 0
n_neighbors = 60
dataset_prefix = "./data/varying_Re"
arbitrary_name = "Re$(Re)_n$(n_neighbors)_symmetricRotation"

data_matrix = Matrix(CSV.read("$(dataset_prefix)/$(arbitrary_name).csv", DataFrame))

#define an equation with variables
eq = (sin.(data_matrix[:,2]).^2.6111207 .- 0.6100381)./(data_matrix[:,1] .+ 0.80143833).^data_matrix[:,1]
#compute mse
eq = data_matrix[:,11] .- eq
mse = sum(eq.^2)/length(eq)
println("MSE: ", mse)
#compute R^2
R2 = 1 - sum(eq.^2)/(sum((data_matrix[:,11].-mean(data_matrix[:,11])).^2).+1e-30)
println("R^2: ", R2)
```
