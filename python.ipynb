```python
import os

# Set your target directory here
folder_path = './GP_data_varying_Re/varying_Re_all_n30_hidden30_dropout_AIFL'

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('val.csv'):
        # Build full original file path
        old_path = os.path.join(folder_path, filename)
        
        # Replace 'val.csv' with 'test.csv'
        new_filename = filename[:-7] + 'test.csv'  # remove 'val.csv', add 'test.csv'
        new_path = os.path.join(folder_path, new_filename)
        
        # Rename the file
        os.rename(old_path, new_path)
```
